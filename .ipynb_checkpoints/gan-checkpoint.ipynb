```python
import pandas
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import random
```

```python
def generate_real():
    real_data = torch.FloatTensor([random.uniform(0.8, 1.0),
                                  random.uniform(0.0, 0.2),
                                  random.uniform(0.8, 1.0),
                                  random.uniform(0.0, 0.2)])
    return real_data

generate_real()
```

```python
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(nn.Linear(4, 3),
                                 nn.Sigmoid(),
                                 nn.Linear(3, 1),
                                 nn.Sigmoid())
        # Cost-Funktion erzeugen
        self.loss_function = nn.MSELoss()
        self.optimiser = torch.optim.SGD(self.parameters(), lr=0.01)
        self.counter = 0
        self.progress = []
    
    def forward(self, inputs):
        # Modell einfach ausführen
        return self.model(inputs)

    def train(self, inputs, targets):
        # Den Ausgang des Netzes berechnen
        outputs = self.forward(inputs)
        # Verlust berechnen
        loss = self.loss_function(outputs, targets)
        
        self.counter += 1
        if(self.counter % 10 == 0):
            self.progress.append(loss.item())
            #print(self.progress)
            if(self.counter % 10000 == 0):
                print("counter = ", self.counter)
        
        # Gradienten nullen, ackpropagation ausführen
        self.optimiser.zero_grad()
        loss.backward()
        self.optimiser.step()

    def plot_progress(self):
        df = pandas.DataFrame(self.progress, columns=['loss'])
        df.plot(ylim=(0, 1.0), figsize=(16, 8), alpha=0.5, marker='.', ms=1,
                grid=True, yticks=(0, 0.25, 0.5))
```

```python
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(nn.Linear(1, 3),
                                 nn.Sigmoid(),
                                 nn.Linear(3, 4),
                                 nn.Sigmoid())
        # Cost-Funktion erzeugen
        self.loss_function = nn.MSELoss()
        self.optimiser = torch.optim.SGD(self.parameters(), lr=0.01)
        self.counter = 0
        self.progress = []
    
    def forward(self, inputs):
        # Modell einfach ausführen
        return self.model(inputs)
```

```python
def generate_random(size):
    random_data = torch.rand(size)
    return random_data
```

```python
D = Discriminator()

for i in range(10000):
    # Echte Daten
    D.train(generate_real(),
           torch.FloatTensor([1.0]))
    # Fake-Daten
    D.train(generate_random(4),
           torch.FloatTensor([0.0]))
    
D.plot_progress()
```

```python
print(D.forward(generate_real()).item())
print(D.forward(generate_random(4)).item())
```
